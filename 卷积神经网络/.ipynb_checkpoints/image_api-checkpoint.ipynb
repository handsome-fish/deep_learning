```python
# resize
# crop
# flip
# brightness & contrast

import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
```

```python
name = './image/cat.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)

sess = tf.Session()
img_decoded_val = sess.run(img_decoded)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# resize
# tf.image.resize_area
# tf.image.resize_bicubic
# tf.image.resize_nearest_neighbour

name = './image/cat.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
# resize函数是针对mini batch的，要求数据四维，是处理多个数据
img_decoded = tf.reshape(img_decoded, [1, 500, 700, 3])

# resize_img = tf.image.resize_bicubic(img_decoded, [1000, 1000])
resize_img = tf.image.resize_nearest_neighbor(img_decoded, [1000, 1000])

sess = tf.Session()
img_decoded_val = sess.run(resize_img)
# 输出用np调整shape为三维
img_decoded_val = img_decoded_val.reshape([1000, 1000, 3])
# imshow 要求数据为int型
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# crop
# tf.image.pad_to_bounding_box
# tf.image.crop_to_bounding_box
# tf.image.random_crop

name = './image/cat.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
# resize函数是针对mini batch的，要求数据四维，是处理多个数据
img_decoded = tf.reshape(img_decoded, [1, 500, 700, 3])

padded_img = tf.image.pad_to_bounding_box(img_decoded, 50, 80, 600, 800)

sess = tf.Session()
img_decoded_val = sess.run(padded_img)
# 输出用np调整shape为三维
img_decoded_val = img_decoded_val.reshape([600, 800, 3])
# imshow 要求数据为int型
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# flip
# tf.image.flip_up_down
# tf.image.flip_left_right
# tf.image.random_flip_up_down
# tf.image.random_flip_left_right

name = './image/cat.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
# resize函数是针对mini batch的，要求数据四维，是处理多个数据
img_decoded = tf.reshape(img_decoded, [1, 500, 700, 3])

flipped_img = tf.image.random_flip_left_right(img_decoded)

sess = tf.Session()
img_decoded_val = sess.run(flipped_img)
# 输出用np调整shape为三维
img_decoded_val = img_decoded_val.reshape([500, 700, 3])
# imshow 要求数据为int型
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```

```python
# brightness & constrast
# tf.image.adjust_brightness
# tf.image.random_brightness
# tf.image.adjust_constrast
# tf.image.random_constrast

name = './image/cat.jpg'
img_string = tf.read_file(name)
img_decoded = tf.image.decode_image(img_string)
# resize函数是针对mini batch的，要求数据四维，是处理多个数据
img_decoded = tf.reshape(img_decoded, [1, 500, 700, 3])

new_img = tf.image.adjust_brightness(img_decoded, +0.5)

sess = tf.Session()
img_decoded_val = sess.run(new_img)
# 输出用np调整shape为三维
img_decoded_val = img_decoded_val.reshape([500, 700, 3])
# imshow 要求数据为int型
img_decoded_val = np.asarray(img_decoded_val, np.uint8)
print(img_decoded_val.shape)

%matplotlib inline
imshow(img_decoded_val)
```
