```python
import os
import sys
import pprint

input_description_file = './dataset/flickr30k/results_20130124.token'
output_vocab_file = './dataset/flickr30k/vocab.txt'
```

```python
def count_vocab(input_description_file):
    '''Generate vocabulary.
    In addition, count distribution of length of sentence 
    and max length of image desciption. 
    '''
    with open(input_description_file, 'r', encoding = 'utf-8') as f:
        lines = f.readlines()
        
    max_length_of_sentence = 0
    length_dict = {}
    vocab_dict = {}
    for line in lines:
        image_id, description = line.strip('\r\t').split('\t')
        sentence = description.strip(' ').split()
        max_length_of_sentence = max(max_length_of_sentence, len(sentence))
        length_dict.setdefault(len(sentence), 0)
        length_dict[len(sentence)] += 1
        
        for word in sentence:
            vocab_dict.setdefault(word, 0)
            vocab_dict[word] += 1
    print(max_length_of_sentence)
    pprint.pprint(length_dict)
    pprint.pprint(vocab_dict)
    return vocab_dict

vocab_dict = count_vocab(input_description_file)
```

```python
sorted_vacab_dict = sorted(vocab_dict.items(), key = lambda d:d[1], reverse = True)
with open(output_vocab_file, 'w', encoding = 'utf-8') as f:
    f.write('<UNK>\t1000000\n')
    for item in sorted_vacab_dict:
        f.write('%s\t%d\n' % item)
```
